```python
import numpy as np
from tensorflow.keras import models, layers
from tensorflow.keras.datasets import imdb

# 데이터 로드 및 전처리
# IMDb 데이터셋 로드 (상위 10,000개의 가장 자주 등장하는 단어만 사용)
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)

# 데이터 벡터화 함수
def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.  # 해당 단어 인덱스에 1 할당
    return results

# 훈련 데이터와 테스트 데이터 벡터화
x_train = vectorize_sequences(train_data)
x_test = vectorize_sequences(test_data)

# 레이블(0 또는 1)은 이미 벡터화되어 있으므로 그대로 사용
y_train = np.asarray(train_labels).astype("float32")
y_test = np.asarray(test_labels).astype("float32")

# 모델 정의
model = models.Sequential([
    layers.Dense(16, activation="relu", input_shape=(10000,)),
    layers.Dense(16, activation="relu"),
    layers.Dense(1, activation="sigmoid")  # 이진 분류를 위해 sigmoid 사용
])

# 모델 컴파일
model.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])

# 모델 훈련
history = model.fit(
    x_train, y_train,
    epochs=20,
    batch_size=512,
    validation_split=0.4
)

# 훈련 결과 출력
import matplotlib.pyplot as plt

def plot_training_history(history):
    plt.figure(figsize=(12, 5))

    # 손실 그래프
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()

    # 정확도 그래프
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

# 훈련 결과 시각화
plot_training_history(history)

# 테스트 데이터 평가
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_acc}")
```

```python
import numpy as np
from tensorflow.keras import models, layers, regularizers
from tensorflow.keras.datasets import imdb
import matplotlib.pyplot as plt

# IMDb 데이터셋 로드 및 전처리
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)

# 데이터 벡터화 함수
def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.  # 해당 단어 인덱스에 1 할당
    return results

# 훈련 데이터와 테스트 데이터 벡터화
x_train = vectorize_sequences(train_data)
x_test = vectorize_sequences(test_data)

# 레이블(0 또는 1)은 이미 벡터화되어 있으므로 그대로 사용
y_train = np.asarray(train_labels).astype("float32")
y_test = np.asarray(test_labels).astype("float32")

# 기본 모델 정의
basic_model = models.Sequential([
    layers.Dense(16, activation="relu", input_shape=(10000,)),
    layers.Dense(16, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])

# L2 규제를 적용한 모델 정의
l2_model = models.Sequential([
    layers.Dense(16, activation="relu", kernel_regularizer=regularizers.l2(0.002), input_shape=(10000,)),
    layers.Dense(16, activation="relu", kernel_regularizer=regularizers.l2(0.002)),
    layers.Dense(1, activation="sigmoid")
])

# 두 모델 컴파일
for model in [basic_model, l2_model]:
    model.compile(optimizer="rmsprop",
                  loss="binary_crossentropy",
                  metrics=["accuracy"])

# 두 모델 훈련
history_basic = basic_model.fit(
    x_train, y_train,
    epochs=20,
    batch_size=512,
    validation_split=0.4,
    verbose=0  # 훈련 로그 생략
)

history_l2 = l2_model.fit(
    x_train, y_train,
    epochs=20,
    batch_size=512,
    validation_split=0.4,
    verbose=0  # 훈련 로그 생략
)

# 규제 효과 시각화
def plot_history_comparison(history_basic, history_l2):
    plt.figure(figsize=(12, 5))

    # 손실 비교
    plt.subplot(1, 2, 1)
    plt.plot(history_basic.history['val_loss'], label='Basic Model (Validation Loss)')
    plt.plot(history_l2.history['val_loss'], label='L2 Regularized Model (Validation Loss)')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Validation Loss Comparison')
    plt.legend()

    # 정확도 비교
    plt.subplot(1, 2, 2)
    plt.plot(history_basic.history['val_accuracy'], label='Basic Model (Validation Accuracy)')
    plt.plot(history_l2.history['val_accuracy'], label='L2 Regularized Model (Validation Accuracy)')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Validation Accuracy Comparison')
    plt.legend()

    plt.tight_layout()
    plt.show()

# 규제 효과 그래프 출력
plot_history_comparison(history_basic, history_l2)

# 테스트 데이터 평가
basic_loss, basic_acc = basic_model.evaluate(x_test, y_test, verbose=0)
l2_loss, l2_acc = l2_model.evaluate(x_test, y_test, verbose=0)

print(f"Basic Model - Test Loss: {basic_loss:.4f}, Test Accuracy: {basic_acc:.4f}")
print(f"L2 Regularized Model - Test Loss: {l2_loss:.4f}, Test Accuracy: {l2_acc:.4f}")
```

```python
import numpy as np
from tensorflow.keras import models, layers, regularizers
from tensorflow.keras.datasets import imdb
import matplotlib.pyplot as plt

# IMDb 데이터셋 로드 및 전처리
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)

# 데이터 벡터화 함수
def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.  # 해당 단어 인덱스에 1 할당
    return results

# 훈련 데이터와 테스트 데이터 벡터화
x_train = vectorize_sequences(train_data)
x_test = vectorize_sequences(test_data)

# 레이블(0 또는 1)은 이미 벡터화되어 있으므로 그대로 사용
y_train = np.asarray(train_labels).astype("float32")
y_test = np.asarray(test_labels).astype("float32")

# 기본 모델 정의
basic_model = models.Sequential([
    layers.Dense(16, activation="relu", input_shape=(10000,)),
    layers.Dense(16, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])

# L2 규제를 적용한 모델 정의
l2_model = models.Sequential([
    layers.Dense(16, activation="relu", kernel_regularizer=regularizers.l2(0.002), input_shape=(10000,)),
    layers.Dense(16, activation="relu", kernel_regularizer=regularizers.l2(0.002)),
    layers.Dense(1, activation="sigmoid")
])

# Dropout을 적용한 모델 정의
dropout_model = models.Sequential([
    layers.Dense(16, activation="relu", input_shape=(10000,)),
    layers.Dropout(0.5),  # Dropout 추가
    layers.Dense(16, activation="relu"),
    layers.Dropout(0.5),  # Dropout 추가
    layers.Dense(1, activation="sigmoid")
])

# 세 모델 컴파일
for model in [basic_model, l2_model, dropout_model]:
    model.compile(optimizer="rmsprop",
                  loss="binary_crossentropy",
                  metrics=["accuracy"])

# 세 모델 훈련
history_basic = basic_model.fit(
    x_train, y_train,
    epochs=20,
    batch_size=512,
    validation_split=0.4,
    verbose=0  # 훈련 로그 생략
)

history_l2 = l2_model.fit(
    x_train, y_train,
    epochs=20,
    batch_size=512,
    validation_split=0.4,
    verbose=0  # 훈련 로그 생략
)

history_dropout = dropout_model.fit(
    x_train, y_train,
    epochs=20,
    batch_size=512,
    validation_split=0.4,
    verbose=0  # 훈련 로그 생략
)

# 규제 효과 시각화
def plot_history_comparison(history_basic, history_l2, history_dropout):
    plt.figure(figsize=(18, 5))

    # 손실 비교
    plt.subplot(1, 2, 1)
    plt.plot(history_basic.history['val_loss'], label='Basic Model (Validation Loss)')
    plt.plot(history_l2.history['val_loss'], label='L2 Regularized Model (Validation Loss)')
    plt.plot(history_dropout.history['val_loss'], label='Dropout Model (Validation Loss)')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Validation Loss Comparison')
    plt.legend()

    # 정확도 비교
    plt.subplot(1, 2, 2)
    plt.plot(history_basic.history['val_accuracy'], label='Basic Model (Validation Accuracy)')
    plt.plot(history_l2.history['val_accuracy'], label='L2 Regularized Model (Validation Accuracy)')
    plt.plot(history_dropout.history['val_accuracy'], label='Dropout Model (Validation Accuracy)')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Validation Accuracy Comparison')
    plt.legend()

    plt.tight_layout()
    plt.show()

# 규제 효과 그래프 출력
plot_history_comparison(history_basic, history_l2, history_dropout)

# 테스트 데이터 평가
basic_loss, basic_acc = basic_model.evaluate(x_test, y_test, verbose=0)
l2_loss, l2_acc = l2_model.evaluate(x_test, y_test, verbose=0)
dropout_loss, dropout_acc = dropout_model.evaluate(x_test, y_test, verbose=0)

print(f"Basic Model - Test Loss: {basic_loss:.4f}, Test Accuracy: {basic_acc:.4f}")
print(f"L2 Regularized Model - Test Loss: {l2_loss:.4f}, Test Accuracy: {l2_acc:.4f}")
print(f"Dropout Model - Test Loss: {dropout_loss:.4f}, Test Accuracy: {dropout_acc:.4f}")
```
